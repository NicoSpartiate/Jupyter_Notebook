```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Ex1a.csv', sep="\t")
df.columns = df.columns.str.replace(' ', '')
df = df[df.columns.drop(list(df.filter(regex='Unnamed')))]

df
```

```python
df['temp']
```

```python

fig, axes = plt.subplots(2,2, sharex=False, sharey=False, figsize=(15,10))
fig.suptitle('Résultats', fontsize=16)
axes = axes.flatten()
axes[0].scatter(df['temp'], df['pH'])
axes[0].set_xlabel('temp')
axes[0].set_ylabel('pH')
axes[0].set_title('pH(temp)')
#axes[0].set_xlim([20, 30])
#axes[0].set_ylim([3, 4])

axes[1].plot(df['temp'], df['Alk'], 'o')
axes[1].set_xlabel('temp')
axes[1].set_ylabel('Alk')
axes[1].set_title('Alk(temp)')

axes[2].plot(df['temp'], df['Alk'], 'r')
axes[2].set_xlabel('temp')
axes[2].set_ylabel('charge')
axes[2].set_title('charge(temp)')

axes[3].plot(df['temp'], df['Alk'])
axes[3].ticklabel_format(axis="y", style="sci", scilimits=(0,0), useMathText=True)
axes[3].set_xlabel('temp')
axes[3].set_ylabel('pct_err')
axes[3].set_title('pct_err(temp)')

plt.show()
```
